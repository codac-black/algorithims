```python
# Brensenham Line Drawing Algorithm with non inverted y-axis
import matplotlib.pyplot as plt
import seaborn as sns

def draw_line(x0, y0, x1, y1):
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    x, y = x0, y0
    p = 2 * dy - dx

    points = [(x, y)]

    while x <= x1:
        if p >= 0:
            y = y + 1 if y1 > y0 else y - 1
            p = p - 2 * dx
        p = p + 2 * dy
        x = x + 1
        points.append((x, y))

    return points

def main():
    try:
        x0, y0 = map(int, input("Enter start point (x0 y0): ").replace(",", " ").split())
        x1, y1 = map(int, input("Enter end point (x1 y1): ").replace(",", " ").split())
    except ValueError:
        print("Invalid input")
        return 0
    except:
        print("Something went wrong")
        return 0

    points = draw_line(x0, y0, x1, y1)

    # Use Seaborn for styling (optional)
    sns.set(style="whitegrid")

    # Plot the points
    x_values, y_values = zip(*points)
    plt.scatter(x_values, y_values, color='red', marker='o')
    plt.title("Bresenham Line Drawing")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

if __name__ == "__main__":
    main()
```

```python
# Bresenham Line Drawing Algorithm with inverted y-axis
import matplotlib.pyplot as plt
import seaborn as sns

def draw_line(x0, y0, x1, y1):
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    x, y = x0, y0
    p = 2 * dy - dx

    points = [(x, y)]

    while x <= x1:
        if p >= 0:
            y = y + 1 if y1 > y0 else y - 1
            p = p - 2 * dx
        p = p + 2 * dy
        x = x + 1
        points.append((x, y))

    return points

def main():
    try:
        x0, y0 = map(int, input("Enter start point (x0 y0): ").replace(",", " ").split())
        x1, y1 = map(int, input("Enter end point (x1 y1): ").replace(",", " ").split())
    except ValueError:
        print("Invalid input")
        return 0
    except:
        print("Something went wrong")
        return 0

    points = draw_line(x0, y0, x1, y1)

    # Use Seaborn for styling (optional)
    sns.set(style="whitegrid")

    # Plot the points
    fig, ax = plt.subplots()
    x_values, y_values = zip(*points)
    ax.scatter(x_values, y_values, color='red', marker='o')
    
    # Customize axis appearance
    ax.set_title("Bresenham Line Drawing")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.invert_yaxis()  # Invert y-axis
    
    plt.show()

if __name__ == "__main__":
    main()
```
