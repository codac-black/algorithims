```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_palette('bright')

# DDA Algorithim
# DDA is a line drawing algorithim which is used to draw a line from one point to another
# it is a basic algorithim which is used in computer graphics
# it is used to draw a line on a square grid

def dda(x1, x2, y1, y2):
    # initialize the starting points
    x, y = (x1, y1)
    # calculate the difference in Xs and Ys
    dx = x2 - x1
    dy = y2 - y1

    # lenght will take the value of the largets value
    length = max(abs(dx), abs(dy))

    x_inc = dx / float(length)
    y_inx = dy / float(length)
    
    # create a list to store x , y cordinates

    x_points = [x]
    x_trun_points = [x]
    y_points = [y] 

    # calculate and store the intermediate points

    for _ in range(length):
        x += x_inc
        y += y_inx
        x_points.append(round(x))
        x_trun_points.append(int(x))
        y_points.append(round(y))

    return x_points, x_trun_points , y_points

# cordinates
x1, y1 = (11,2)
x2, y2 = (4, 25)

# call the DDA Algorithim to get the points
x_points, x_trun_points, y_points = dda(x1, x2, y1, y2)


# plot the points
plt.figure(figsize=(6, 6))
plt.plot(x_points, y_points, 'g')
plt.plot(x_trun_points, y_points, 'b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('DDA Algorithim')
# plt.grid(True)
sns.set_style('whitegrid')
sns.set_palette('bright')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np

def dda_algorithm(x1, y1, z1, x2, y2, z2):
    points = []
    dx = x2 - x1
    dy = y2 - y1
    dz = z2 - z1
    steps = max(abs(dx), abs(dy), abs(dz))
    x_increment = dx / steps
    y_increment = dy / steps
    z_increment = dz / steps

    x, y, z = x1, y1, z1
    for _ in range(steps + 1):
        points.append((round(x), round(y), round(z)))
        x += x_increment
        y += y_increment
        z += z_increment

    return points

# Define the line coordinates in 3D space
x1, y1, z1 = 11,2, 5
x2, y2, z2 = 4, 25, 12

# Calculate the line points using DDA algorithm
line_points = dda_algorithm(x1, y1, z1, x2, y2, z2)

# Set the width and height
width, height = 12, 10

# Create a 3D plot to visualize the line
fig = plt.figure(figsize=(width, height))
ax = fig.add_subplot(111, projection='3d')

# Plot the line points in 3D
x_values, y_values, z_values = zip(*line_points)
ax.plot(x_values, y_values, z_values, marker='o', color='green')

# Set Seaborn style
sns.set(style="whitegrid")

# Show the 3D plot
plt.show()
```
